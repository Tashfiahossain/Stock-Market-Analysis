```python

#Calculate and plot the daily percentage change in closing prices(For ACI company)


data['daily percentage change']=data['Close'].pct_change()*100


plt.figure(figsize=(12,5))
plt.plot(data['Date'],data['daily percentage change'])
plt.xlabel('Date')
plt.ylabel(f'{company2}daily percentage change')
plt.title(f'{company2}daily percentage change in Closing Price')
```

```python

#Investigate the presence of any trends(daily return) or seasonality in the stock prices for ACI.

data['Trend']=data['Close']-data['Open']
plt.figure(figsize=(12,5))
plt.plot(data['Date'],data['Trend'])
plt.xlabel('Date')
plt.ylabel(f'{company2}Stock Trends')
plt.title(f'{company2}trends(daily return) or seasonality in the stock prices')
```

```python

#Investigate the presence of any trends or seasonality in the stock prices for ACI.
#(price trends by plotting the stock's open, close,high, and low prices over a specific time period.)

plt.figure(figsize=(12,5))
plt.plot(data['Date'],data['Close'],color='blue',label= f'{company2} Closing Price')
plt.plot(data['Date'],data['Open'],color='red',label= f'{company2} Opening Price')
plt.plot(data['Date'],data['High'],color='green',label= f'{company2} High Price')
plt.plot(data['Date'],data['Low'],color='yellow',label= f'{company2} Low Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title(f'{company2} Prices over time')

plt.legend()

plt.show()
```

```python

#Apply moving averages to smooth the time series data in 15/30 day intervals against the original graph(For ACI company).


data['15-day Rolling avg']= data['Close'].rolling(window=15).mean()

plt.figure(figsize=(12,5))
plt.plot(data['Date'],data['Close'],color='blue',label= f'{company2} Closing Price')
plt.plot(data['Date'],data['15-day Rolling avg'],color='red',label= f'{company2} 15-day Rolling avg of Closing Price')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title(f'{company2} moving average in 15 day intervals against the original graph')

plt.legend()

plt.show()
```

```python

#Identify stocks with unusually high trading volume on certain days.

mean_volume = df['Volume'].mean()
std_volume = df['Volume'].std()

df['Volume Z-Score'] = (df['Volume'] - mean_volume) / std_volume
threshold = 2 
unusual_volume_days = df[df['Volume Z-Score'] > threshold]

unusual_volume_days.nlargest(10,'Volume Z-Score')
```

```python

#Calculate the correlation matrix between the 'Open' & 'High', 'Low' &'Close' prices( For ACI company).


data[['Open', 'High']].corr()
data[['Low', 'Close']].corr()
```

```python

#Create a heatmap to visualize the correlations using the seaborn package (For ACI company).



x=data[['Open', 'High']].corr()
plt.subplot(1, 2, 1) 
sns.heatmap(x,annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title(' Open & High')


y=data[['Low', 'Close']].corr()
plt.subplot(1, 2, 2) 
sns.heatmap(y,annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title(' Low & Close')

plt.tight_layout() 
```
